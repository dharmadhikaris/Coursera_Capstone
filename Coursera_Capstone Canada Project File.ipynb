```python
### Get request for data to convert and converted into text, set to url variable ###
url = requests.get("https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M").text
```

```python
### Set 'scrape' function from BeautifulSoup ###
scrape = BeautifulSoup(url, 'lxml')
```

```python
### Assign 'table' and 'table_rows' functions ###
table = scrape.find("table")
table_rows = table.tbody.find_all("tr") 
```

```python
### Construct for loop to append'Not Assigned' bouroughs and neighborhoods ###
res = []
for tr in table_rows:
    td = tr.find_all("td")
    row = [tr.text for tr in td]

    if row != [] and row[1] != "Not assigned":
        if "Not assigned" in row [2]:
            row [2] = row[1]
        res.append(row)
### Tailor data frame to only include relevant columns ###
cd = pd.DataFrame(res, columns = ["PostalCode", "Bourough", "Neighborhood"])
cd.head()
```

```python
### Clean data frame ###

cd["Neighborhood"] = cd["Neighborhood"].str.replace("\n", "")
```

```python
cd.head()
```

```python
cd["PostalCode"] = cd["PostalCode"].str.replace("\n", "")
```

```python
cd["Bourough"] = cd["Bourough"].str.replace("\n", "")
```
